```python
def get_data(tech_list,start,end,tick):

    data_list = {}
    for stock in tech_list:
        data_list[stock] = yf.download(stock, start, end,interval=tick)
    
    return data_list
```

```python
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)
tick = '1d'

tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']

data_list = get_data(tech_list,start,end,tick)
```
