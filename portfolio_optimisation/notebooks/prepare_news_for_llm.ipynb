```python
import pandas as pd
import zipfile
input_file = 'news_sp_500_best_25.csv'
output_file = 'news_sp_500_best_25_weekly.csv'
path = '../../datasets/'
```

```python
df = pd.read_csv(f'{path}{input_file}')
```

```python
examples_df = df.head()
```

```python
    # end date in the same format
examples_df['created'] = pd.to_datetime(examples_df['created'])
examples_df.head()
```

```python

start_date = '2010-09-02'  # start date in the format Year-Month-Day
end_date = '2010-09-04'
mask = (examples_df['created'] >= start_date) & (examples_df['created'] <= end_date)
filtered_df = examples_df.loc[mask]
filtered_df.head()
```

```python
df['created'] = pd.to_datetime(df['created'], errors='coerce')
df = df.dropna(subset=['created'])
```

```python
df.sort_values(by=['created'], inplace=True, ascending=True)
df.head()
```

```python
import tiktoken
encoding = tiktoken.get_encoding("cl100k_base")

encoding = tiktoken.encoding_for_model("gpt-4-1106-preview")

def num_tokens_from_string(string, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    if type(string) is not str:
        return 0
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
df['body_tokens'] = df['body'].apply(lambda x: num_tokens_from_string(x, "cl100k_base"))
```

```python
df['title'] = df['title'].astype(str)
mask = df['title'].str.startswith("Stocks That Hit 52-Week")
filtered_df = df[~mask]
```

```python
filtered_df.nlargest(10, 'body_tokens')
```

```python
# 2009-07-27 08:38:00-04:00
# 2023-10-30 14:20:03-04:00
filtered_df.set_index('created', inplace=True)
```

```python
weekly_data = {}
import ast
import itertools
from textblob import TextBlob

def extract_names(row):
    if pd.isna(row):
        return []
    dict_list = ast.literal_eval(row)
    return [d['name'] for d in dict_list]

def concatenate_list_of_list(lol):
    return list(itertools.chain.from_iterable(lol))

def tokenize_and_contatenate(list_of_text):
    # print(list_of_text)
    return ' '.join([' '.join(TextBlob(str(text)).words) for text in list_of_text])

for period_start, group in filtered_df.resample('W'):
    index = period_start.strftime('%Y-%m-%d')
    best_news = group.nlargest(10, 'body_tokens')
    if sum(best_news['body_tokens']) > 32000:
        for i in range(1, 10):
            best_news = group.nlargest(10-i, 'body_tokens')
            if sum(best_news['body_tokens']) < 32000:
                break
    period_object = {
        'body': tokenize_and_contatenate(best_news['body'].tolist()),
        'title': tokenize_and_contatenate(best_news['title'].tolist()),
        'stocks': set(concatenate_list_of_list(best_news['stocks'].apply(extract_names).tolist()))
    }
    weekly_data[index] = period_object
```

```python
items = weekly_data.items()
keys_to_delete = []
for key, value in items:
    if not value['title']:
        keys_to_delete.append(key)

for key in keys_to_delete:
    del weekly_data[key]
```

```python
df_merged = pd.DataFrame.from_dict(weekly_data, orient='index')
df_merged.head()
```

```python
df_merged.to_csv(f'{path}{output_file}')
```
