```python
df['created'] = pd.to_datetime(df['created'], errors='coerce')
df = df.dropna(subset=['created'])
```

```python
df.sort_values(by=['created'], inplace=True, ascending=True)
df.head()
```

```python
# 2009-07-27 08:38:00-04:00
# 2023-10-30 14:20:03-04:00
df.set_index('created', inplace=True)
```

```python
weekly_data = {}
import ast
import itertools
from textblob import TextBlob

def extract_names(row):
    if pd.isna(row):
        return []
    dict_list = ast.literal_eval(row)
    return [d['name'] for d in dict_list]

def concatenate_list_of_list(lol):
    return list(itertools.chain.from_iterable(lol))

def tokenize_and_contatenate(list_of_text):
    # print(list_of_text)
    return ' '.join([' '.join(TextBlob(str(text)).words) for text in list_of_text])
for period_start, group in df.resample('W'):
    index = period_start.strftime('%Y-%m-%d')
    period_object = {
        'body': tokenize_and_contatenate(group['body'].tolist()),
        'title': tokenize_and_contatenate(group['title'].tolist()),
        'stocks': set(concatenate_list_of_list(group['stocks'].apply(extract_names).tolist()))
    }
    weekly_data[index] = period_object
```

```python
weekly_data.keys()
```

```python
weekly_data['2009-08-02']
```

```python
items = weekly_data.items()
keys_to_delete = []
for key, value in items:
    if not value['title']:
        keys_to_delete.append(key)

for key in keys_to_delete:
    del weekly_data[key]
```

```python
df_merged = pd.DataFrame.from_dict(weekly_data, orient='index')
df_merged.head()
```

```python
df_merged.to_csv(f'{path}{output_file}')
```
